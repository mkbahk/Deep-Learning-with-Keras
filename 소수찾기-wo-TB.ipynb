1~7을 학습해 다음 수 8, 9, 10을 예측하는 예제


```
from itertools import permutations
import math
import numpy as np
from tensorflow import keras
```

Input Data용 배열생성

```
x = [0, 1]
y = [0, 0]
```

```
def check(n):
    k = math.sqrt(n)
    if n < 2: 
        return 0
    #end of if

    for i in range(2, int(k)+1):
        if n % i == 0:
            return 0
    #end of for

    return 1
#end of def

for p in range(2, 65536, 1):
  x.append(p)
  y.append(check(p))
#end of for

#print(x)
#print(y)
```

훈련용과 테스트용 데이타셋으로 나누기: x는 입력값, y는 정답

```
x_train = x[:65500]
y_train = y[:65500]
x_test = x[65500:]
y_test = y[65500:]
```

모델정의


```
model = keras.Sequential()
model.add(keras.layers.Dense(1000, input_dim = 1, activation='sigmoid'))  #input layer
model.add(keras.layers.Dense(2000, activation='sigmoid'))  # 1개의 hidden layer
model.add(keras.layers.Dense(1000, activation='sigmoid'))  # 1개의 hidden layer
model.add(keras.layers.Dense(20000, activation='sigmoid'))  # 1개의 hidden layer
model.add(keras.layers.Dense(1000, activation='sigmoid'))  # 1개의 hidden layer
model.add(keras.layers.Dense(2000, activation='sigmoid'))  # 1개의 hidden layer
model.add(keras.layers.Dense(1000, activation='sigmoid'))  # 1개의 hidden layer
model.add(keras.layers.Dense(1, activation='sigmoid')) # output layer
```

모델 컴파일

```
#model.compile(optimizer=keras.optimizers.SGD(learning_rate=0.1), loss="categorical_crossentropy", metrics=['accuracy'])
#model.compile(optimizer='adam', loss='mse', metrics=['accuracy'])
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

모델 훈련

```
model.fit(x=x_train, y=y_train, epochs=50, batch_size=100)
```

예측하기

```
y_predict = model.predict(x=x_test)
print(np.round(y_predict))
```
