```
import numpy as np
```

```
A = np.array([[10, 20, 30], [40, 50, 60]])
print(A.shape)
print(A)
```

```
row_add = np.array([70, 80, 90]).reshape(1, 3)
print(row_add.shape)
print(row_add)
```

```
column_add = np.array([1000, 2000]).reshape(2, 1)
print(column_add.shape)
print(column_add)
```

```
B = np.concatenate((A, row_add), axis=0)
print(B)
```

```
C = np.concatenate((A, column_add), axis=1)
print(C)
```
