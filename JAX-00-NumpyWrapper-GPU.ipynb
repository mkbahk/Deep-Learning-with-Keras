```python
import numpy as np
x = np.ones((5000, 5000))
y = np.arange(5000)

%timeit z = np.sin(x) + np.cos(y)  
```

```python
import jax.numpy as jnp
x = jnp.ones((5000, 5000))
y = jnp.arange(5000)

%timeit z = jnp.sin(x) + jnp.cos(y) 
```

```python
from jax import jit
import tensorflow as tf

def fn(x, y):
  z = np.sin(x)
  w = np.cos(y)
  return z + w
### end of def:

@jit
def fn_jit(x, y):
  z = jnp.sin(x)
  w = jnp.cos(y)
  return z + w
### end of def:

@tf.function
def fn_tf2(x, y):
  z = tf.sin(x)
  w = tf.cos(y)
  return z + w
###end of def:
```

```python
x = np.ones((5000, 5000))
y = np.ones((5000, 5000))

%timeit fn(x, y)
```

```python
jx = jnp.ones((5000, 5000))
jy = jnp.ones((5000, 5000))

%timeit fn_jit(jx, jy)
```

```python
tx = tf.ones((5000, 5000))
ty = tf.ones((5000, 5000))

%timeit fn_tf2(tx, ty)
```
