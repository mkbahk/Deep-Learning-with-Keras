```
def naive_relu(x):
  assert len(x.shape) == 2 #x는 2D 넘파이 배열입니다.

  x = x.copy() #입력 텐서 자체를 바꾸지 않도록 복사합니다.

  for i in range(x.shape[0]):
    for j in range(x.shape[1]):
      x[i, j] = max(x[i, j], 0)
  return x
#end of def

def naive_add(x, y):
  assert len(x.shape) == 2 # x와 y는 넘파이 배열입니다.
  assert len(y.shape) == 2
  assert x.shape == y.shape

  x = x.copy() #입력텐서 자체를 바꾸지 않습니다.

  for i in range(x.shape[0]):
    for j in range(x.shape[1]):
      x[i, j] += y[i, j]
  return x
#end of def

def naive_add_matrix_and_vector(x, y):
  assert len(x.shape) == 2 # x는 2D 넘파이 배열입니다.
  assert len(y.shape) == 1 # y는 넘파이 1D벡터입니다.
  assert x.shape[1] == y.shape[0]

  x = x.copy() #입력 텐서 자체를 바꾸지 않토록 복사합니ㅏㄷ.

  for i in range(x.shape[0]):
    for j in range(x.shape[1]):
      x[i, j] += y[j]
  return x
#end of def
```

```
import numpy as np
```

```
x = np.array([[5, 78, 2, 34, 0],
             [6, 79, 3, 35, 1],
             [7, 80, 4, 36, 2]])
print(naive_relu(x))
```

```
y = np.array([[5, 78, 2, 34, -1],
             [6, 79, 3, 35, -1],
             [7, 80, 4, 36, 2]])
print(naive_add(x, y))
```

```
z = x + y
print(z)
```

```
z = np.maximum(z, 0.) # 원소별 렐루 함수
print(z)
```

```
x = np.random.random((64, 3, 32, 10)) # x는 (64, 3, 32, 10) 크기의 랜덤 텐서입니다.
y = np.random.random((32, 10)) # Y는 (32, 10) 크기의 램덤 텐서입니다.

z = np.maximum(x, y)
print(z)
```
