```
A = np.array([1, 2, 3])
B = np.array([4, 5, 6])

# vector A, B 값 출력
print("A ==", A, "\n", "B ==", B)

# vector A, B 형상 출력 => shape
print("A.shape ==", A.shape, ", B.shape ==", B.shape)

# vector A, B 차원 출력 => ndim
print("A.ndim ==", A.ndim, ", B.ndim ==", B.ndim)
```

```
# vector 산술연산

print("A + B ==", A + B)
print("A - B ==", A - B)
print("A * B ==", A * B)
print("A / B ==", A / B)
```

```
A = np.array([[1, 2, 3], [4, 5, 6]])
B = np.array([[-1, -2, -3], [-4, -5, -6]])

# vector A, B 값 출력
print("A ==", A, "\n", "B ==", B)

# vector A, B 형상 출력 => shape
print("A.shape ==", A.shape, ", B.shape ==", B.shape)

# vector A, B 차원 출력 => ndim
print("A.ndim ==", A.ndim, ", B.ndim ==", B.ndim)
```

```
# matrix 산술연산

print("A + B ==", A + B)
print("A - B ==", A - B)
print("A * B ==", A * B)
print("A / B ==", A / B)
```

```
# 형변환
C = np.array([1, 2, 3])

# vector C형상 출력 => shape
print("C.shape ==", C.shape, "벡터")
print(C)

# vector를 (1, 3) 행렬로 형 변환
C = C.reshape(1, 3)
print("C.shape ==", C.shape, "행력")
print(C)
```
